```python
import numpy as np
import matplotlib.pyplot as plt


# from thm 16:
def P(d_A, d_B, N_A, epsilon):
    a = np.log(2)
    b = (
        8
        * N_A
        * np.log(
            (40 * (2 * np.sqrt(2) + 1) ** 2 * N_A**2 * d_A**2 * d_B**2) / (epsilon**2)
        )
    )
    c = (-d_A * d_B * epsilon**2) / (18 * np.pi**3 * (4 * np.sqrt(2) + 2) ** 2)
    return np.exp(a + b + c)
```

```python
# plot for fixed d_B, increasing d_A
# params:
d_B = 2
N_As = np.arange(10, 15)
d_As = 2**N_As
P_values = [P(2**N_A, d_B, N_A, 0.5) for N_A in N_As]
# plot
plt.figure(figsize=(8, 5))
plt.plot(N_As, P_values, marker="o")
plt.xlabel("d_A (2^N_A)")
plt.ylabel("P(d_A, d_B=2, N_A, epsilon=0.5)")
plt.title("Probability Bound vs d_A (fixed d_B=2, epsilon=0.5)")
plt.xscale("log", base=2)
plt.grid(True)
plt.show()
```
