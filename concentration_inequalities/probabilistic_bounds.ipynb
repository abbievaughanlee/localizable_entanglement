```python
import numpy as np
import matplotlib.pyplot as plt
# from thm 16:
def P(d_A, d_B, N_A, epsilon):
    term1 = np.log(2) * 8 * N_A * np.log((40 * (2 * np.sqrt(2) + 1)**2 * N_A**2 * d_A**2 * d_B**2) / (epsilon**2))
    term2 = (d_A * d_B * epsilon**2) / (18 * np.pi**3 * (4 * np.sqrt(2) + 2)**2)
    return np.exp(term1 - term2)
```

```python
# plot for fixed d_A, increasing d_B:
N_A = 1
d_A = 2**N_A  # d_A = 2
epsilon = 0.5

d_B_values = np.arange(1, 1001)  # d_B from 1 to 20
P_values = [P(d_A, d_B, N_A, epsilon) for d_B in d_B_values]

plt.figure(figsize=(8,5))
plt.plot(d_B_values, P_values, marker='o')
plt.xlabel('d_B')
plt.ylabel('P(d_A=2, d_B, N_A=1, epsilon=0.5)')
plt.title('Probability Bound vs d_B (fixed d_A=2, N_A=1, epsilon=0.5)')
plt.grid(True)
plt.show()
```
